# Importe de paquetes

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Funciones para nuestros objetos b√°sicos

```python
def comp_y(y0, vy0, t, a = -9.8):
    result = y0 + vy0*t + a*t**2/2
    return(result)
```

```python
comp_y(1, 5, 0.5)
```

```python
def comp_x(x0, vx0, t):
    result = x0 + vx0*t
    return(result)
```

```python
comp_x(0, 8, 1)
```

```python
t = np.arange(0,5,0.1)  # Devuelve un arreglo [0, 0.1, 0.2, 0.3, ..., 4.9, 5]
y = list(map(lambda t: comp_y(5000, 30, t), t))
```

```python
#help(plt)
```

```python
%matplotlib inline
plt.plot(t, y)
plt.title("Trayectoria de la componente y en el tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Altura (m)")
plt.show()
```

```python
def trayectoria(x0, y0, vx0, vy0, t, a = -9.8):
    y = comp_y(y0, vy0, t, a)
    x = comp_x(x0, vx0, t)
    return(x,y)
```

```python
t = np.arange(0,5,0.1)
r = lambda t: trayectoria(0, 15, 2, 25, t)
r = list(map(r, t))
r
```

```python
rx = [x for x,y in r]
ry = [y for x,y in r]
```

```python
plt.plot(rx, ry)
plt.title("Trayectoria del proyectil")
plt.xlabel("Recorrido horizontal (m)")
plt.ylabel("Altura (m)")
plt.show()
```
